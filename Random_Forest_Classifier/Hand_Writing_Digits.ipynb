```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the datasets from sklearn.datasets libraries
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
#Creating the Pandas DataFrame
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
df["target"] = digits.target
df
```

```python
df["name"] = df["target"].apply(lambda x : digits.target_names[x])
df
```

```python
#Extracting independent and Dependent variables
x = df.drop(["target","name"],axis=1)
y = df["target"]
```

```python
x
```

```python
y
```

```python
#splitting the dataset into training and test datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#training the dataset to RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=40)
model.fit(x_train,y_train)
```

```python
#test the accuracy
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
#classification report
from sklearn.metrics import classification_report
cl = classification_report(y_test,y_predicted)
print(cl)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
#visualizing confusion matrix
import seaborn as sns
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
